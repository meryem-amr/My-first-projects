# Data preprocessing

## 1)Data transformation:

```python
#changing the date into datetime type
df["Date"] = pd.to_datetime(df["Date"], format="%d-%m-%Y")
```

```python
#making sure that type of column Date has changed
df.info()
```

The type of Date has changed into datetime

```python
#adding columns to the dataframe
df['month'] = df.Date.dt.month
df['year'] = df.Date.dt.year
df['week'] = df['Date'].dt.isocalendar().week
```

```python
#partitionong the year into 2 semesters and adding column semester to the dataframe
first_semester = [1,2,3,4,5,6]
second_semester = [7,8,9,10,11,12]
df['semester'] = ['First' if month in first_semester else 'Second' for month in df['month']]
```
